```python
def step(z):
    return 1 if z>0 else 0
```

```python
intercept_, coef_ = Perceptron(X, y)
```

```python
intercept_, coef_
```

```python
# y = -w1/w2*x -b/w2 or y = mx+c
m = -(coef_[0]/coef_[1])
c = -(intercept_/coef_[1])
```

```python
x_input = np.linspace(-3, 3, 100)
y_input = m*x_input + c
```

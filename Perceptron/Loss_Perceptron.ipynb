```python
from sklearn.datasets import make_classification
import numpy as np
X, y = make_classification(n_samples=100, n_features=2, n_informative=1, n_redundant=0, n_classes=2, n_clusters_per_class=1, random_state=41, hypercube=False, class_sep=15)
```

```python
X.shape, y.shape
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8,5))
plt.scatter(X[:,0], X[:, 1], c=y, cmap='winter', s=80)
plt.show()
```

```python
def Perceptron(X, y):
    # Initialize weights
    w1=w2=b=1
    # Learning rate
    lr = 0.1
    epochs = 1000
    
    for j in range(epochs):
        for i in range(X.shape[0]):
            # Check condition
            z = w1*X[i][0] + w2*X[i][1] + b
            
            if z*y[i] < 0:
                w1 = w1 + lr*y[i]*X[i][0]
                w2 = w2 + lr*y[i]*X[i][1]
                b = b + lr*y[i]

    return w1, w2, b
```

```python
w1, w2, b = Perceptron(X, y)
```

```python
w1, w2, b
```

```python
m = -w1/w2
c = -b/w2

m, c
```

```python
x_input = np.linspace(-3, 3, 100)
y_input = m*x_input + c
```

```python
plt.figure(figsize=(8,5))
plt.plot(x_input, y_input, color='red', linewidth=2)
plt.scatter(X[:, 0], X[:, 1], c = y, cmap='winter', s=70)
plt.ylim(-3, 2)
```
